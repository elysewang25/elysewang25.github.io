```python
import pandas as pd
import requests
import io
import bqplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('https://query.data.world/s/ly237ztepnjxemwmftzkmb6v4gbdgd')
```

```python
df
```

```python
is50 = pd.to_numeric(df['World_Rank'], errors = 100)<=50
df50 = df[is50]
df50
```

```python
df50_country = df50.groupby(['Country','Year'])['Total_Score'].count()
df50_country
```

```python
new_df50 = pd.DataFrame(df50_country).reset_index()
new_df50.columns = ['Country','Year','Count']
new_df50
```

```python
plt.figure(figsize=(20,10))
for i in np.unique(new_df50['Country']):
    filterdf = (new_df50['Country'] == i)
    plt.plot(new_df50[filterdf]['Year'],new_df50[filterdf]['Count'], marker='.', linewidth=1.0, label=i)
plt.xlabel("Year")
plt.ylabel("# of University")
plt.legend()
```
