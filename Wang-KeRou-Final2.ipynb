```python
import pandas as pd
import requests
import io
import bqplot
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

```python
# get the data from https://query.data.world/s/ly237ztepnjxemwmftzkmb6v4gbdgd
df = pd.read_excel('https://query.data.world/s/ly237ztepnjxemwmftzkmb6v4gbdgd')
```

```python
# collect data with only top 50 university
is50 = pd.to_numeric(df['World_Rank'], errors = 100)<=50
df50 = df[is50]
```

```python
names = np.unique(df50['Country'], return_inverse=True)[1]

namen = np.unique(df50['Country'], return_inverse=True)
pd.DataFrame(data = namen[0], columns = ['Country'])
```

```python
# give each unique country a numeric id
names = np.unique(df50['Country'], return_inverse=True)[1]
new_df50 = pd.DataFrame(df50)
new_df50['country_id'] = names
new_df50['Weight'] = (names+1)/(names+1)
new_df50
```

```python
nlong = 15
nlat =14
long_bins = np.linspace(0,15, nlong+1)
lat_bins = np.linspace(0, 59, nlat+1)
hist2d, long_edges, lat_edges = np.histogram2d(x = new_df50['country_id'], y = new_df50['World_Rank'],
                                               weights=new_df50['Weight'],
                                              bins=[long_bins, lat_bins])
hist2d.shape, long_edges.shape, lat_edges.shape
hist2d
```

```python
#long_centers = [(long_edges[i]+long_edges[i+1])*0.5 for i in range(len(long_edges)-1)]
lat_centers = [(lat_edges[i]+lat_edges[i+1])*0.5 for i in range(len(lat_edges)-1)]
#long_centers, lat_centers
lat_centers
```

```python
# (I) Heatmap
# scales
col_sc = bqplot.ColorScale(scheme='RdPu')
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# axes
c_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical',side='right')
x_ax = bqplot.Axis(scale=x_sc, label='Country ID')
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label='World Rank')

# marks: heatmap
heat_map = bqplot.GridHeatMap(color=hist2d.T, row=lat_centers, column = long_edges,
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'},
                             selected_style={'opacity':1.0},
                             unselected_stye={'opacity':1.0})
```

```python
# (II) Bar plot of the numbers of university
x_scl = bqplot.LinearScale()
y_scl = bqplot.LinearScale()
ax_xcl = bqplot.Axis(label='Year', scale=x_scl)
ax_ycl = bqplot.Axis(label='Number of Universities', scale=y_scl, 
                    orientation = 'vertical', side='left')
j,i = 0,0
longs = [long_edges[j], long_edges[j+1]]
lats = [lat_edges[i],lat_edges[i+1]]
region_mask = ( (new_df50['World_Rank']>lats[0]) & (new_df50['World_Rank']<= lats[1]) &\
              (new_df50['country_id']>=longs[0]) & (new_df50['country_id']< longs[1]))

dur, dur_edges = np.histogram(new_df50['Year'][region_mask],
                              weights=new_df50['Weight'][region_mask],
                             bins=10)
# calculate center of our bins
#dur_centers = [(dur_edges[i]+dur_edges[i+1])*0.5 for i in range(len(dur_edges)-1)]
# make histogram by hand using bqplot.Bars
duration_hist = bqplot.Bars(x=dur_edges, y=dur,
                           scales={'x':x_scl,'y':y_scl})
fig_dur = bqplot.Figure(marks=[duration_hist], axes=[ax_xcl,ax_ycl])
```

```python
# (V) Our interactions
# lets start building our dashboard
mySelectedLabel = ipywidgets.Label()
def get_data_value(change):
    j,i = change['owner'].selected[0]
    #print(i,j)
    v = hist2d[i,j]
    mySelectedLabel.value = 'Numbers of University = ' + str(v)
    longs = [long_edges[i], long_edges[i+1]]
    lats = [lat_edges[j],lat_edges[j+1]]
    region_mask = ( (new_df50['World_Rank']>lats[0]) & (new_df50['World_Rank']<= lats[1]) &\
              (new_df50['country_id']>=longs[0]) & (new_df50['country_id']< longs[1]))
    dur, dur_edges = np.histogram(new_df50['Year'][region_mask], 
                                  
                             bins=10)
    #dur_centers = [(dur_edges[k]+dur_edges[k+1])*0.5 for k in range(len(dur_edges)-1)]
    duration_hist.x = dur_edges
    duration_hist.y = dur
```

```python
# link this label to changes in our histogram
heat_map.observe(get_data_value, 'selected')

fig = bqplot.Figure(marks=[heat_map],axes=[c_ax,y_ax,x_ax])
#fig
ipywidgets.VBox([mySelectedLabel,ipywidgets.HBox([fig,fig_dur])])
```
